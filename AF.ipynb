**Riassunto di Spazi Vettoriali**

Un **spazio vettoriale** su un campo *F* (come i numeri reali *R* o i numeri complessi *C*) è un insieme *V* con due operazioni:

*   Somma tra vettori: Per ogni coppia di vettori *x*, *y* in *V*, la loro somma *x* + *y* è anche un vettore in *V*.
*   Prodotto scalare-vettore: Per ogni scalare *α* in *F* e ogni vettore *x* in *V*, il loro prodotto *αx* è un vettore in *V*.

Queste operazioni devono soddisfare alcune proprietà fondamentali come l'associatività, la commutatività della somma, l'esistenza di un vettore nullo e di un inverso additivo per ogni vettore, e le proprietà distributive.

**Esempi di spazi vettoriali:**

*   Lo spazio *F<sup>n</sup>* delle *n*-uple di elementi di *F*.
*   L'insieme *C(I)* delle funzioni continue da un intervallo *I* in *F*.
*   L'insieme *C(Ω)* delle funzioni continue da uno spazio topologico *Ω* in *F*.
*   L'insieme *P* di tutti i polinomi con coefficienti in *F*.
*   Il prodotto diretto *V* × *W* di due spazi vettoriali *V* e *W*.
*   L'insieme *F(S, V)* delle funzioni da un insieme *S* a uno spazio vettoriale *V*.

**Sottospazi Vettoriali**

Un **sottospazio vettoriale** di *V* è un sottoinsieme *U* di *V* che è esso stesso uno spazio vettoriale con le stesse operazioni di *V*. Un sottoinsieme *U* di *V* è un sottospazio vettoriale se e solo se è chiuso rispetto alle combinazioni lineari, ovvero se per ogni *x*, *y* in *U* e per ogni *α*, *β* in *F*, il vettore *αx* + *βy* è anche in *U*.

**Esempi di sottospazi vettoriali:**

*   Per ogni spazio vettoriale *V*, {0} e *V* sono sottospazi vettoriali di *V*.
*   Le rette passanti per l'origine sono sottospazi vettoriali del piano *R<sup>2</sup>*.
*   L'insieme *P<sub>d</sub>* dei polinomi di grado al più *d* è un sottospazio vettoriale di *P*.

**Combinazioni Lineari e Indipendenza Lineare**

Dati *k* vettori *v<sub>1</sub>*, ..., *v<sub>k</sub>* in uno spazio vettoriale *V*, una **combinazione lineare** di questi vettori è un vettore della forma *α<sub>1</sub>v<sub>1</sub>* + ... + *α<sub>k</sub>v<sub>k</sub>*, dove *α<sub>1</sub>*, ..., *α<sub>k</sub>* sono scalari. I vettori *v<sub>1</sub>*, ..., *v<sub>k</sub>* si dicono **linearmente indipendenti** se l'unica combinazione lineare che dà come risultato il vettore nullo è quella con tutti i coefficienti *α<sub>1</sub>* = ... = *α<sub>k</sub>* = 0. Altrimenti, si dicono **linearmente dipendenti**.

**Basi e Dimensione**

Una **base** di uno spazio vettoriale *V* è un insieme di vettori linearmente indipendenti che generano *V* tramite combinazioni lineari. La **dimensione** di *V*, denotata con dim *V*, è la cardinalità (numero di elementi) di una sua base.

**Applicazioni Lineari**

Un'**applicazione lineare** tra due spazi vettoriali *V* e *W* è una funzione *T* : *V* → *W* che preserva le operazioni di somma e prodotto scalare-vettore: *T(αx + βy) = αTx + βTy* per ogni *x, y* in *V* e *α, β* in *F*.

**Proprietà importanti:**

*   Il **nucleo** di *T*, denotato con Ker *T*, è l'insieme dei vettori *x* in *V* tali che *Tx = 0*.
*   L'**immagine** di *T*, denotata con Im *T*, è l'insieme dei vettori *y* in *W* tali che esiste un *x* in *V* con *Tx = y*.
*   Un'applicazione lineare è iniettiva se e solo se il suo nucleo è {0}.
*   Un'applicazione lineare è suriettiva se e solo se la sua immagine è uguale a *W*.

**Matrici e Applicazioni Lineari**

Ogni applicazione lineare tra spazi vettoriali di dimensione finita può essere rappresentata da una matrice. La matrice associata a un'applicazione lineare dipende dalla scelta delle basi per gli spazi vettoriali di partenza e di arrivo.

### Spazi Metrici e Topologia

**Definizione di spazio metrico:** Uno spazio metrico è un insieme M dotato di una funzione d: M × M → [0,∞) che soddisfa le seguenti proprietà:
* **Proprietà di separazione:** Per ogni x, y ∈ M, d(x, y) = 0 se e solo se x = y.
* **Simmetria:** Per ogni x, y ∈ M, d(x, y) = d(y, x).
* **Disuguaglianza triangolare:** Per ogni x, y, z ∈ M, d(x, z) ≤ d(x, y) + d(y, z).

La funzione d è detta distanza o metrica su M.

**Esempi di spazi metrici:**

* Lo spazio euclideo **R<sup>n</sup>** con la metrica euclidea d2(x, y) = √∑<sub>j=1</sub><sup>n</sup> |xj − yj|<sup>2</sup>.
* Lo spazio **C<sup>n</sup>** con la metrica euclidea d2(x, y) = √∑<sub>j=1</sub><sup>n</sup> |xj − yj|<sup>2</sup>.
* Lo spazio delle funzioni continue **C[a, b]** con la metrica dell'estremo superiore d∞(f, g) := sup<sub>t∈[a,b]</sub> |f(t)−g(t)| = max<sub>t∈[a,b]</sub> |f(t)−g(t)|. 

**Successioni in spazi metrici:**

* Una successione (xn)<sub>n∈N</sub> in uno spazio metrico (M, d) **converge** a un punto x ∈ M se per ogni ε > 0 esiste N ∈ N tale che d(xn, x) < ε per ogni n > N.
* Una successione (xn)<sub>n∈N</sub> in uno spazio metrico (M, d) è di **Cauchy** se per ogni ε > 0 esiste N ∈ N tale che d(xn, xm) < ε per ogni n, m > N. 

**Spazi metrici completi:** Uno spazio metrico (M, d) è **completo** se ogni successione di Cauchy a valori in M converge a un punto di M.

**Esempi di spazi metrici completi:**

* **R<sup>n</sup>** con la metrica euclidea.
* **C<sup>n</sup>** con la metrica euclidea.
* Lo spazio delle funzioni continue e limitate **C<sub>b</sub>(S, F)** con la metrica dell'estremo superiore.

**Convergenza puntuale e uniforme:**

* Una successione di funzioni (f<sub>n</sub>)<sub>n∈N</sub> converge **puntualmente** a una funzione f se lim<sub>n→∞</sub> f<sub>n</sub>(t) = f(t) per ogni t nell'insieme di definizione. 
* Una successione di funzioni (f<sub>n</sub>)<sub>n∈N</sub> converge **uniformemente** a una funzione f se lim<sub>n→∞</sub> d∞(f<sub>n</sub>, f) = 0.

**Topologia in spazi metrici:**

* Un sottoinsieme U di uno spazio metrico (M, d) è un **intorno** di un punto x ∈ M se esiste r > 0 tale che B(x, r) ⊆ U, dove B(x, r) è la **palla aperta** di centro x e raggio r.
* Un sottoinsieme A di M è **aperto** se è un intorno di ogni suo punto.
* Un sottoinsieme C di M è **chiuso** se il suo complementare M \ C è aperto.

**Compattezza in spazi metrici:**

* Un sottoinsieme K di uno spazio metrico (M, d) è **compatto** se ogni successione a valori in K ha una sottosuccessione che converge a un punto di K.
* Un sottoinsieme E di M è **relativamente compatto** se la sua chiusura E̅ è compatta.

**Teorema di Heine-Borel:** In **R<sup>n</sup>** (o **C<sup>n</sup>**), un sottoinsieme è compatto se e solo se è chiuso e limitato.

**Continuità in spazi metrici:** Una funzione f: M → N tra spazi metrici è **continua** in un punto x ∈ M se per ogni ε > 0 esiste δ > 0 tale che d<sub>N</sub>(f(x), f(x')) < ε per ogni x' ∈ M con d<sub>M</sub>(x, x') < δ.

**Teorema di Heine-Cantor:** Se M è uno spazio metrico compatto e f: M → N è una funzione continua, allora f è **uniformemente continua**.

**Funzioni lipschitziane:** Una funzione f: M → N tra spazi metrici è **lipschitziana** se esiste una costante L ≥ 0 tale che d<sub>N</sub>(f(x), f(x')) ≤ L d<sub>M</sub>(x, x') per ogni x, x' ∈ M. Una funzione lipschitziana è uniformemente continua.

**Separabilità:** Uno spazio metrico (M, d) è **separabile** se ha un sottoinsieme denso e numerabile.

**Esempi di spazi metrici separabili:**

* **R<sup>n</sup>** con la metrica euclidea.
* **C<sup>n</sup>** con la metrica euclidea.
* Lo spazio delle funzioni continue **C[a, b]** con la metrica dell'estremo superiore.

## Spazi Normati e Spazi di Banach

**Definizione:** Uno spazio vettoriale $X$ su un campo $\mathbb{F}$ (dove $\mathbb{F}$ sono i numeri reali $\mathbb{R}$ o i numeri complessi $\mathbb{C}$) con una funzione $ \|\cdot\| : X \rightarrow [0, \infty)$ è chiamato spazio normato se per ogni $x, y \in X$ e $\alpha \in \mathbb{F}$ sono soddisfatte le seguenti condizioni:

*   $\|x\| = 0$ se e solo se $x = 0$ (proprietà di separazione);
*   $\|\alpha x\| = |\alpha| \|x\|$ (1-omogeneità);
*   $\|x + y\| \le \|x\| + \|y\|$ (disuguaglianza triangolare).

**Definizione**: La distanza $d: X \times X \rightarrow [0, \infty)$ indotta dalla norma $\|\cdot\|$ è definita come $d(x, y) = \|x - y\|$ per ogni $x, y \in X$.

Questa definizione implica che ogni spazio normato è anche uno spazio metrico con la distanza indotta dalla norma.

### Esempi di spazi normati

*   Per ogni $n \in \mathbb{N}^+$ e $p \in [1, \infty]$, lo spazio vettoriale $(\mathbb{F}^n, \|\cdot\|_p)$ è uno spazio normato dove:
    
    $\|x\|_p =  \begin{cases} (\sum_{j=1}^n |x_j|^p)^{1/p} & \text{se } p < \infty, \\ \max\{|x_j| : j = 1, ..., n\} & \text{se } p = \infty \end{cases}$  per ogni $x \in \mathbb{F}^n$.
    
    La distanza indotta in questo caso è $d_p$. In particolare, per $p=2$, ($\mathbb{F}^n, d_2$) è uno spazio metrico completo e separabile, il che significa che $(\mathbb{F}^n, \|\cdot\|_2)$ è uno spazio di Banach separabile.
*   Se $M$ è uno spazio metrico compatto, allora $(\mathcal{C}(\mathbb{F}(M)), \|\cdot\|_\infty)$ è uno spazio normato, dove $\|\cdot\|_\infty$ è la norma dell'estremo superiore:
    
    $\|f\|_\infty = \sup_{x \in M} |f(x)|$ per ogni $f \in \mathcal{C}(\mathbb{F}(M))$.
    
    La distanza indotta è $d_\infty$ e poiché $(\mathcal{C}(\mathbb{F}(M)), d_\infty)$ è uno spazio metrico completo, $(\mathcal{C}(\mathbb{F}(M)), \|\cdot\|_\infty)$ è uno spazio di Banach.
*   Se $(X, \|\cdot\|)$ è uno spazio normato e $Y$ è un sottospazio vettoriale di $X$, allora la restrizione a $Y$ della norma $\|\cdot\|$ rimane una norma su $Y$. Questa restrizione è spesso chiamata *norma indotta*.
*   Se $(X, \|\cdot\|_X)$ e $(Y, \|\cdot\|_Y)$ sono spazi normati, il prodotto diretto $X \times Y$ è uno spazio normato con la norma prodotto:
    
    $\|(x, y)\|_{X \times Y} = \max\{\|x\|_X, \|y\|_Y\}$ per ogni $(x, y) \in X \times Y$.

### Proprietà degli Spazi Normati

Dato uno spazio normato $(X, \|\cdot\|)$:

*   La norma $\|\cdot\| : X \rightarrow [0, \infty)$ è 1-lipschitziana: $|\|x\| - \|y\|| \le \|x - y\|$ per ogni $x, y \in X$.
*   Le operazioni di somma e prodotto scalare-vettore sono continue. In altre parole:
    *   Se $x_n \rightarrow x$ e $y_n \rightarrow y$ in $X$, allora $x_n + y_n \rightarrow x + y$ in $X$.
    *   Se $x_n \rightarrow x$ in $X$ e $\alpha_n \rightarrow \alpha$ in $\mathbb{F}$, allora $\alpha_n x_n \rightarrow \alpha x$ in $X$.
*   Per ogni $x \in X$ e $r > 0$, le palle $B(x, r)$ e $\overline{B}(x, r)$ sono convesse.
*   Un sottoinsieme $E$ di $X$ è limitato se e solo se $\sup_{x \in E} \|x\| < \infty$.
*   Una successione $(x_n)_{n \in \mathbb{N}}$ a valori in $X$ è limitata se e solo se $\sup_{n \in \mathbb{N}} \|x_n\| < \infty$.

### Norme Equivalenti

Due norme $\|\cdot\|$ e $\|\cdot\|'$ su uno spazio vettoriale $X$ su un campo $\mathbb{F}$ sono dette equivalenti se esistono costanti $A, B \in (0, \infty)$ tali che:

$\|x\| \le A \|x\|'$ e $\|x\|' \le B \|x\|$ per ogni $x \in X$.

L'equivalenza tra norme è una relazione di equivalenza, il che significa che è riflessiva, simmetrica e transitiva.

Due norme $\|\cdot\|$ e $\|\cdot\|'$ su $X$ sono equivalenti se e solo se  $id_X : (X, \|\cdot\|) \rightarrow (X, \|\cdot\|')$ è bilipschitziana (cioè lipschitziana con inversa lipschitziana).

Ecco alcune conseguenze dell'equivalenza di due norme $\|\cdot\|$ e $\|\cdot\|'$ su uno spazio $X$:

*   Le due norme inducono la stessa topologia su $X$. Di conseguenza, convergenza di successioni, continuità di funzioni, e le proprietà di essere aperto, chiuso, compatto o denso per i sottoinsiemi rimangono invariate se si sostituisce una norma con l'altra.
*   Un sottoinsieme $E \subseteq X$ è limitato in $(X, \|\cdot\|)$ se e solo se è limitato in $(X, \|\cdot\|')$.
*   Una successione $(x_n)_n$ è di Cauchy in $(X, \|\cdot\|)$ se e solo se è di Cauchy in $(X, \|\cdot\|')$.

### Equivalenza delle Norme, Completezza e Chiusura

In uno spazio vettoriale $X$ su un campo $\mathbb{F}$ di dimensione finita, tutte le norme sono equivalenti. 

Di conseguenza, in uno spazio vettoriale $X$ di dimensione finita, ogni norma induce la stessa topologia e:

*   $(X, \|\cdot\|)$ è uno spazio di Banach separabile.
*   Ogni sottoinsieme chiuso e limitato di $X$ è compatto.

Consideriamo uno spazio normato $(X, \|\cdot\|)$ e un suo sottospazio vettoriale $Y$. Valgono le seguenti proprietà:

*   Se $(Y, \|\cdot\|)$ è completo, allora $Y$ è chiuso in $X$.
*   Se $\text{dim}Y < \infty$, allora $Y$ è chiuso in $X$.
*   Se $(X, \|\cdot\|)$ è uno spazio di Banach, allora $Y$ è chiuso in $X$ se e solo se $(Y, \|\cdot\|)$ è completo.

### Serie Convergenti e Assolutamente Convergenti

Sia $(X, \|\cdot\|)$ uno spazio normato, $(x_n)_{n \in \mathbb{N}}$ una successione a valori in $X$ e $x \in X$.

*   Si dice che la serie $\sum_{n=0}^\infty x_n$ converge a $x$ se $\lim_{N \rightarrow \infty} \sum_{n=0}^N x_n = x$. In questo caso, si scrive anche $\sum_{n=0}^\infty x_n = x$ e $x$ è detta somma della serie.
*   Si dice che la serie $\sum_{n=0}^\infty x_n$ converge assolutamente se $\sum_{n=0}^\infty \|x_n\| < \infty$.

Se $X$ è uno spazio di Banach e la serie $\sum_{n=0}^\infty x_n$ converge assolutamente, allora converge a qualche $x \in X$.

Inoltre, se la serie $\sum_{n=0}^\infty x_n$ converge assolutamente a $x$, allora la convergenza è incondizionata, ovvero $\sum_{n=0}^\infty x_{\sigma(n)} = x$ per ogni biiezione $\sigma : \mathbb{N} \rightarrow \mathbb{N}$. In altre parole, la somma della serie non dipende dall'ordine degli addendi. 

**Nota:**  Mentre la convergenza incondizionata implica la convergenza assoluta per serie a valori reali, in spazi normati di dimensione infinita, possono esistere serie convergenti incondizionatamente ma non assolutamente.

### Compattezza in spazi normati

**Lemma di Riesz:** Se $(X, \|\cdot\|)$ è uno spazio normato e $Y \subseteq X$ è un sottospazio chiuso proprio (cioè $Y \neq X$), allora esiste un $w \in X$ tale che $\|w\| = 1$ e $d(w, Y) \ge 1/2$.

Una conseguenza del Lemma di Riesz è che in uno spazio normato $(X, \|\cdot\|)$ di dimensione infinita, $B(0, 1)$ e $S(0, 1) := \{x \in X : \|x\| = 1\}$ non sono compatti.

Le seguenti affermazioni sono equivalenti per uno spazio normato $(X, \|\cdot\|)$:

*   Ogni successione limitata in $X$ ha una sottosuccessione convergente.
*   Tutti i sottoinsiemi chiusi e limitati di $X$ sono compatti in $X$.
*   $B(0, 1)$ è compatta in $X$.
*   $\text{dim}X < \infty$.

### Separabilità in spazi normati

Uno spazio normato $(X, \|\cdot\|)$ è separabile se e solo se esiste un sottoinsieme $E \subseteq X$ al più numerabile tale che $X = \text{span}E$.

**Teorema di Stone-Weierstrass:** Per ogni intervallo chiuso e limitato $[a, b]$ in $\mathbb{R}$ e ogni funzione $f \in \mathcal{C}[a, b]$, esiste una successione di polinomi $(p_n)_{n \in \mathbb{N}}$ in $P$ tale che $p_n|_{[a, b]} \Rightarrow f$.

Di conseguenza, lo spazio normato $(\mathcal{C}[a, b], \|\cdot\|_\infty)$ è separabile.

## Riassunto degli Spazi Lp

Ecco un riassunto degli spazi Lp basato sulle informazioni fornite:

**Spazi di Successioni**

*   **ℓ∞**: Lo spazio vettoriale delle successioni limitate a valori in F (R o C). La norma è definita come  **∥x∥∞ = supk∈N |xk |**.
*   **c0**: Il sottospazio vettoriale di ℓ∞ contenente le successioni infinitesime (convergenti a 0).
*   **c00**: Il sottospazio vettoriale di ℓ∞ contenente le successioni definitivamente nulle.
*   **ℓp**: Per p ∈ [1,∞), ℓp è lo spazio vettoriale delle successioni p-sommabili a valori in F, con norma **∥x∥p = (∑∞ k=0 |xk |p )1/p**.

**Proprietà Importanti:**

*   (ℓ∞, ∥ · ∥∞) e (c0, ∥ · ∥∞) sono spazi di Banach.
*   c00 ⊆ ℓp ⊆ c0 per ogni p ∈ [1,∞). Queste inclusioni sono proprie.
*   c00 è denso in (ℓp, ∥ · ∥p) e ℓp è denso in (c0, ∥ · ∥∞).
*   (ℓp, ∥ · ∥p) è uno spazio di Banach per ogni p ∈ [1,∞].

**Spazi di Funzioni Misurabili**

*   **L0**: L'insieme delle funzioni misurabili da uno spazio di misura (M,M, µ) a F.
*   **Lp**: Per p ∈ [1,∞], Lp(M,M, µ) è l'insieme delle funzioni misurabili f: M → F tali che **∥f ∥p < ∞**, dove ∥f ∥p è definita come:
    *   **(∫M |f |p dµ )1/p** per p ∈ [1,∞)
    *   **ess sup |f |** per p = ∞.
*   **Lp(M,M, µ)**: Per ogni p ∈ [1,∞], Lp(M,M, µ) è lo spazio quoziente L p(M,M, µ)/∼, dove ∼ è la relazione di equivalenza che identifica funzioni uguali quasi ovunque.

**Proprietà Importanti:**

*   Lp(M,M, µ) è uno spazio vettoriale con le operazioni di somma e prodotto per scalare definite sulle classi di equivalenza.
*   ∥ · ∥p è una norma su Lp(M,M, µ) per ogni p ∈ [1,∞].
*   Se (M,M, µ) = (N,P(N),#), allora Lp = ℓp.
*   C[a, b] può essere visto come un sottospazio di L∞(a, b) con la norma indotta.
*   C[a, b] è un sottospazio chiuso proprio di L∞(a, b).
*   Se 1 ≤ p ≤ q ≤ ∞ e (M,M, µ) è uno spazio di misura finito, allora Lq(M) ⊆ Lp(M).
*   Cc(a, b), lo spazio delle funzioni continue con supporto compatto in (a, b), è denso in Lp(a, b) per ogni p ∈ [1,∞).
*   Lp(a, b) è separabile per ogni p ∈ [1,∞).

**Teoremi di Convergenza:**

*   **Teorema di Convergenza Monotona**: Se fn : M → [0,∞] sono funzioni misurabili tali che fn ≤ fn+1 quasi ovunque e fn → f puntualmente quasi ovunque, allora f è misurabile e ∫M f dµ = limn→∞ ∫M fn dµ.
*   **Teorema di Convergenza Dominata**: Se fn ∈ L1(M) convergono puntualmente quasi ovunque a una funzione f e esiste una funzione g ∈ L1(M) tale che |fn| ≤ g quasi ovunque per ogni n, allora f ∈ L1(M) e ∫M f dµ = limn→∞ ∫M fn dµ.
*   **Corollario**: Il Teorema di Convergenza Dominata vale anche per fn ∈ Lp(M) con p ∈ [1,∞), a patto che la funzione dominante g appartenga a Lp(M).

## Riassunto Dei Prodotti Scalari e Spazi Pre-Hilbertiani

**Definizione di Prodotto Scalare:**

Un prodotto scalare su uno spazio vettoriale H (reale o complesso) è una funzione che prende due vettori in H e restituisce uno scalare (reale o complesso, rispettivamente), soddisfacendo le seguenti proprietà:

*   **Linearità nella prima variabile:** ⟨αx + βy, z⟩ = α⟨x, z⟩ + β⟨y, z⟩.
*   **Simmetria (per spazi vettoriali reali) o Hermitiana (per spazi vettoriali complessi):**
    *   Reale: ⟨x, y⟩ = ⟨y, x⟩.
    *   Complesso: ⟨x, y⟩ = ⟨y, x⟩ (coniugato complesso).
*   **Definita Positiva:** ⟨x, x⟩ ≥ 0, e ⟨x, x⟩ = 0 se e solo se x = 0.

**Proprietà Addizionali:**

*   Un prodotto scalare è bilineare in uno spazio vettoriale reale (lineare in entrambe le variabili) e sesquilineare in uno spazio vettoriale complesso (lineare nella prima variabile e antilineare nella seconda).
*   Un prodotto scalare su H è una forma bilineare simmetrica definita positiva su H se H è uno spazio vettoriale reale e una forma sesquilineare hermitiana definita positiva su H se H è uno spazio vettoriale complesso.

**Spazi Pre-Hilbertiani:**

*   Uno spazio vettoriale H dotato di un prodotto scalare è chiamato spazio pre-hilbertiano (o spazio con prodotto scalare).

**Norma Indotta da un Prodotto Scalare:**

*   In uno spazio pre-hilbertiano (H, ⟨·, ·⟩), la norma indotta dal prodotto scalare è definita come: **||x|| = √⟨x, x⟩**.
*   La norma indotta è effettivamente una norma su H.
*   La disuguaglianza di Cauchy-Schwarz afferma che |⟨x, y⟩| ≤ ||x|| ||y|| per ogni x, y in H.

**Spazi di Hilbert:**

*   Uno spazio di Hilbert è uno spazio pre-hilbertiano completo (rispetto alla norma indotta dal prodotto scalare).

**Esempi:**

*   Esempi di spazi di Hilbert includono:
    *   Fn con il prodotto scalare euclideo.
    *   ℓ2 con il prodotto scalare ⟨x, y⟩ = ∑∞ k=0 xkyk.
    *   L2(M,M, µ) con il prodotto scalare ⟨f, g⟩ = ∫M f g dµ.
*   Esempi di spazi pre-hilbertiani che non sono spazi di Hilbert includono:
    *   c00 con il prodotto scalare indotto da ℓ2.
    *   C[a, b] con il prodotto scalare integrale indotto da L2(a, b).

**Proprietà Aggiuntive del Prodotto Scalare e della Norma Indotta:**

*   L'identità di polarizzazione esprime il prodotto scalare in termini della norma indotta.
*   L'identità del parallelogramma mette in relazione la norma di una somma e di una differenza con le norme dei singoli vettori.
*   Il prodotto scalare ⟨·, ·⟩: H × H → F è una funzione continua.
*   L'identità del parallelogramma è una condizione necessaria e sufficiente affinché una norma sia indotta da un prodotto scalare.

Questo riassunto fornisce una panoramica dei prodotti scalari, degli spazi pre-hilbertiani e degli spazi di Hilbert, insieme alle loro proprietà ed esempi.

## Riassunto degli Operatori Lineari Limitati

Questo riassunto si concentra sugli operatori lineari limitati tra spazi normati, come discusso in "06\_operatorilimitati\_handout.pdf".

**Operatori Lineari Limitati**

*   Un operatore lineare T da uno spazio vettoriale X a uno spazio vettoriale Y è una funzione lineare che mappa elementi da X a Y: T: X → Y. Se X = Y, T è chiamato operatore lineare su X.
*   Dato che X e Y sono spazi normati, un operatore lineare T: X → Y è **limitato** se esiste una costante C ≥ 0 tale che la norma di Tx in Y è al massimo C volte la norma di x in X per ogni x in X.
*   La **norma operatoriale** di T, denotata con ||T||<sub>op</sub>, è l'infimum di tutte queste costanti C.
*   Un operatore lineare è limitato se e solo se la sua norma operatoriale è finita.

**Caratterizzazioni ed Esempi**

Il documento fornisce diverse caratterizzazioni equivalenti per gli operatori lineari limitati:

*   **Lipschitziano:** Esiste una costante K ≥ 0 tale che ||Tx - Ty||<sub>Y</sub> ≤ K||x - y||<sub>X</sub> per ogni x, y in X.
*   **Uniformemente Continuo:** Per ogni ε > 0 esiste un δ > 0 tale che ||Tx - Ty||<sub>Y</sub> < ε ogni volta che ||x - y||<sub>X</sub> < δ.
*   **Continuo:** T è continuo in ogni punto di X.
*   **Continuo in 0:** Per ogni ε > 0 esiste un δ > 0 tale che ||Tx||<sub>Y</sub> < ε ogni volta che ||x||<sub>X</sub> < δ.

Vengono forniti diversi esempi di operatori limitati e illimitati, inclusi:

*   **Operatore Identità:** L'operatore identità su uno spazio normato X è limitato con norma operatoriale 1 (se X ≠ {0}).
*   **Operatori di Valutazione:** Per un spazio metrico compatto M, l'operatore di valutazione V<sub>p</sub>: C(M) → F, dove V<sub>p</sub>(f) = f(p), è limitato con norma operatoriale 1.
*   **Operatori Integrali:** Dati due intervalli chiusi e limitati [a, b] e [c, d], e una funzione continua K su [a, b] × [c, d], l'operatore integrale T<sub>K</sub>: C[c, d] → C[a, b] definito da (T<sub>K</sub>f)(x) = ∫<sub>c</sub><sup>d</sup> K(x, y)f(y) dy è limitato.
*   **Operatori di Moltiplicazione:** Per una sequenza limitata w, l'operatore di moltiplicazione D<sub>w</sub>: ℓ<sup>2</sup> → ℓ<sup>2</sup> definito da D<sub>w</sub>(x) = (w<sub>k</sub>x<sub>k</sub>)<sub>k∈ℕ</sub> è limitato.
*   **Proiezioni:** Date le proiezioni π<sub>X</sub>: X × Y → X e π<sub>Y</sub>: X × Y → Y da uno spazio prodotto X × Y ai suoi fattori, questi operatori sono limitati.

**Proprietà degli Operatori Limitati**

Il documento continua esplorando importanti proprietà degli operatori lineari limitati:

*   **Spazio Vettoriale Normato:** L'insieme di tutti gli operatori lineari limitati da X a Y, denotato con B(X, Y), forma uno spazio vettoriale normato con la norma operatoriale.
*   **Chiusura sotto Composizione:** La composizione di operatori lineari limitati è ancora limitata, e la norma operatoriale è submoltiplicativa: ||ST||<sub>op</sub> ≤ ||S||<sub>op</sub>||T||<sub>op</sub>.
*   **Nuclei e Grafici Chiusi:** Il nucleo di un operatore lineare limitato è un sottospazio chiuso, e il suo grafico è un sottospazio chiuso dello spazio prodotto corrispondente.
*   **Estensioni a Spazi di Banach:** Se Y è uno spazio di Banach e D è un sottospazio denso di X, allora ogni operatore lineare limitato da D a Y può essere esteso unicamente a un operatore lineare limitato da X a Y, preservando la sua norma operatoriale.

**Isometrie Lineari e Isomorfismi**

Il documento introduce anche concetti speciali di operatori lineari limitati:

*   **Isometria Lineare:** Un operatore lineare che preserva la norma, cioè ||Tx||<sub>Y</sub> = ||x||<sub>X</sub> per ogni x in X.
*   **Isomorfismo Isometrico:** Un'isometria lineare che è anche biiettiva.
*   **Isomorfismo:** Un operatore lineare limitato biiettivo con un inverso limitato.

**Teoremi Importanti**

Il documento presenta diversi teoremi cruciali riguardanti gli operatori lineari limitati:

*   **Teorema dell'Estensione:** Questo teorema afferma che un operatore lineare limitato definito su un sottospazio denso di uno spazio normato può essere esteso unicamente a tutto lo spazio, preservando la sua norma, a condizione che lo spazio di codominio sia completo (cioè uno spazio di Banach).
*   **Principio dell'Uniforme Limitatezza (Teorema di Banach-Steinhaus):** Questo teorema fornisce una condizione sufficiente affinché una famiglia di operatori lineari limitati sia uniformemente limitata in norma.
*   **Teorema del Grafico Chiuso:** Questo teorema fornisce una condizione necessaria e sufficiente affinché un operatore lineare sia limitato in termini della chiusura del suo grafico.
*   **Teorema dell'Isomorfismo di Banach:** Questo teorema afferma che un operatore lineare limitato biiettivo tra spazi di Banach ha un inverso limitato, stabilendo che è un isomorfismo.
*   **Teorema di Coercività e Invertibilità:** Questo teorema fornisce una caratterizzazione per l'invertibilità di un operatore lineare limitato basata sulla sua coercività in norma e sull'avere un'immagine densa.
*   **Criterio di Invertibilità di Neumann:** Questo teorema fornisce una condizione sufficiente per l'invertibilità di un operatore della forma I - T, dove I è l'operatore identità e ||T||<sub>op</sub> < 1.

Questi concetti e teoremi forniscono una solida base per comprendere gli operatori lineari limitati e le loro proprietà negli spazi normati.

## Analisi Funzionale: Spazio Duale e Funzionali Lineari Continui

Questo riassunto si basa su "07_duale_handout.pdf" e illustra i concetti di spazio duale, funzionali lineari continui e alcuni importanti teoremi come quello di Riesz-Frechet, di Lax-Milgram e di Hahn-Banach.

**Spazio Duale:**

* Dato uno spazio vettoriale X su un campo F, un **funzionale lineare** su X è una mappa lineare da X a F. Lo spazio vettoriale di tutti i funzionali lineari su X è chiamato **duale algebrico** di X e viene indicato con  L(X, F).
* Se X è uno spazio **normato**, il suo **duale** (o **duale topologico**) X' è lo spazio B(X, F) di tutti i **funzionali lineari continui** su X.  X' è a sua volta uno spazio normato con la norma operatoriale.
* **Importanti osservazioni:**
    * Se X ha dimensione finita, il duale algebrico e il duale topologico coincidono.
    * Se X è lo spazio euclideo Fn, i suoi elementi sono vettori colonna (matrici n x 1), mentre gli elementi del suo duale sono vettori riga (matrici 1 x n).
* **Esempi di funzionali lineari continui:** 
    * L'operatore di valutazione su C(M), lo spazio delle funzioni continue su uno spazio metrico compatto M.
    * La mappa ⟨⋅, y⟩ in uno spazio pre-hilbertiano H per ogni y in H.
    * La mappa φy da ℓp a F definita come φy(x) = $\sum_{n=0}^{∞} x_n y_n$, dove p e q sono esponenti coniugati.
    * La mappa φg da Lp(M,M, µ) a F definita come  φg(f) = ∫_M fg dµ, dove p e q sono esponenti coniugati e (M,M, µ) è uno spazio di misura.

**Duale di uno Spazio di Hilbert:**

* Il **teorema di rappresentazione di Riesz-Frechet** afferma che per ogni spazio di Hilbert H su F, esiste una **isometria antilineare suriettiva** Φ da H a H'.
* **Conseguenza**: Lo spazio di Hilbert L2(M,M, µ) è **isometricamente isomorfo** al suo duale.

**Duale degli spazi ℓp e Lp:**

* **ℓp**: Se p e q sono esponenti coniugati, esiste una **isometria lineare** Ψ da ℓq a (ℓp)'.  Se p è finito, Ψ è anche un **isomorfismo isometrico**.
* **Lp**:  Se p e q sono esponenti coniugati e (M,M, µ) è uno spazio di misura **σ-finito**, allora esiste una **isometria lineare** Ψ da Lq(M) a Lp(M)'.  Se p è finito, Ψ è un **isomorfismo isometrico**.

**Forme Sesquilineari e Operatori Lineari:**

* **Forme sesquilineari** sono mappe da X × Y a F che sono lineari nel primo argomento e anilineari nel secondo.
* Esiste una **corrispondenza biunivoca** tra le forme sesquilineari continue su uno spazio normato X e uno spazio di Hilbert H e gli operatori lineari limitati da X a H.
* Il **teorema di Lax-Milgram** stabilisce che, dato uno spazio di Hilbert H e una forma sesquilineare continua e **coerciva** F su H, per ogni funzionale lineare continuo φ su H, esiste un unico elemento y in H tale che φ(x) = F(x, y) per ogni x in H.

**Teorema di Hahn-Banach ed Estensioni di Funzionali:**

* Il **teorema di Hahn-Banach** garantisce l'esistenza di un'estensione lineare continua di un funzionale lineare continuo definito su un sottospazio vettoriale di uno spazio normato all'intero spazio, preservando la sua norma. 
* **Conseguenze:**
    * Per ogni elemento non nullo in uno spazio normato, esiste un funzionale lineare continuo con norma 1 che assume il valore della norma dell'elemento in quel punto.
    * I funzionali lineari continui separano i punti di uno spazio normato non banale.
    * Se il duale di uno spazio normato è separabile, allora lo spazio stesso è separabile.
* **Osservazione:** ℓ∞ non è separabile, mentre ℓp (per p finito) e c0 sono separabili. Di conseguenza, il duale di ℓ∞ non è isomorfo a ℓ1.

## Riepilogo Concetti Chiave da "08_biduale_handout.pdf"

Questo riassunto si basa su "08_biduale_handout.pdf" del Professor Alessio Martini per il corso di Analisi Funzionale al Politecnico di Torino per l'anno accademico 2023/2024.

**Biduale e Spazi Riflessivi:**

* Il biduale (X′′) di uno spazio normato X è il duale del suo spazio duale (X′).
* L'**immersione canonica** J: X → X′′ è un'isometria lineare che collega uno spazio normato al suo biduale.
* Uno spazio normato X è **riflessivo** se l'immersione canonica J è suriettiva, ovvero se l'immagine di J coincide con l'intero spazio X′′.
    * Spazi riflessivi sono completi (di Banach) e isomorfi al loro biduale (X ∼= X′′).
    * L'isomorfismo tra X e X′′ non è sufficiente per garantire la riflessività di X.
    * Spazi di Hilbert, Lp(M,M, µ) con 1 < p < ∞ (M σ-finito), e ℓp con 1 < p < ∞ sono riflessivi.
    * Spazi di dimensione finita sono sempre riflessivi.
    * Spazi come c0, ℓ1, ℓ∞, L1(I), e L∞(I) (I intervallo di misura di Lebesgue positiva) non sono riflessivi.

**Convergenza Debole:**

* In uno spazio normato X, una successione (xn) converge debolmente a x (xn ⇀ x) se φ(xn) converge a φ(x) per ogni φ ∈ X′.
* La convergenza in norma implica la convergenza debole, ma non viceversa in generale.
* In spazi di dimensione finita, convergenza debole e convergenza in norma sono equivalenti.

**Proprietà della Convergenza Debole:**

* Il limite debole è unico.
* Una successione debolmente convergente è limitata.
* In uno spazio di Hilbert, xn ⇀ x se e solo se ⟨xn, y⟩ converge a ⟨x, y⟩ per ogni y ∈ H.
* Caratterizzazioni specifiche della convergenza debole sono fornite per spazi Lp, ℓp e c0.

**Convergenza Debole* sul Duale:**

* In uno spazio di Banach X, una successione (φn) in X′ converge debolmente* a φ (φn ∗⇀ φ) se φn(x) converge a φ(x) per ogni x ∈ X.
* La convergenza debole implica la convergenza debole*, ma non viceversa in generale.
* Caratterizzazioni specifiche della convergenza debole* sono fornite per spazi L∞(M), ℓ∞ e ℓ1.

**Teorema di Banach-Alaoglu:**

* In uno spazio di Banach separabile, ogni successione limitata nel duale ammette una sottosuccessione convergente debolmente*.

**Conseguenze del Teorema di Banach-Alaoglu:**

* In uno spazio di Banach separabile e riflessivo, ogni successione limitata ammette una sottosuccessione debolmente convergente.
* Un funzionale F definito su uno spazio di Banach riflessivo e separabile, con opportune proprietà, soddisfa una proprietà di semicontinuità inferiore rispetto alla convergenza debole.

## Riepilogo Aggiunto di un Operatore e Spettro

### Aggiunto di un Operatore Limitato

Dato un operatore lineare limitato A tra due spazi di Hilbert H1 e H2, l'**aggiunto** di A, denotato con A*, è l'unico operatore lineare limitato da H2 a H1 che soddisfa la seguente relazione:

⟨Ax, y⟩H2 = ⟨x, A\*y⟩H1 per ogni x ∈ H1 e y ∈ H2.

**Proprietà importanti dell'aggiunto:**

* **Involutività:** (A\*)* = A
* **Anti-linearità:** (αA + βB)\* = αA\* + βB\* 
* **Anti-moltiplicatività:** (BA)\* = A\*B\*
* **Norma dell'aggiunto:** ∥A\*A∥op = ∥A∥²op
* **Aggiunto di un isomorfismo:** Se A è un isomorfismo, anche A\* lo è e (A\*)⁻¹ = (A⁻¹)\*
* **Nucleo e immagine:** Ker(A\*) = (ImA)⊥ e ImA = Ker(A\*)⊥

### Operatori Autoaggiunti, Unitari e Normali

Un operatore T ∈ B(H) è detto:

* **autoaggiunto** se T\* = T
* **normale** se T\*T = TT\*
* **unitario** se T\*T = TT\* = idH

**Relazioni tra i tipi di operatori:**

T autoaggiunto ⇒ T normale ⇐ T unitario

**Proprietà degli operatori:**

* Un operatore T è unitario se e solo se è un isomorfismo isometrico.
* Un operatore T è autoaggiunto se e solo se ⟨Tx, x⟩ è reale per ogni x ∈ H (assumendo F = C).
* Un operatore normale T preserva la norma (∥Tx∥H = ∥T\*x∥H) e ha lo stesso nucleo del suo aggiunto (KerT = Ker(T\*)).

### Spettro di un Operatore

Lo **spettro** di un operatore lineare limitato T ∈ B(H), denotato con σ(T), è l'insieme dei numeri complessi λ tali che T - λI non è invertibile in B(H), dove I è l'operatore identità.

**Lo spettro si può decomporre in tre sottoinsiemi:**

* **spettro puntuale** σp(T): contiene gli autovalori di T. 
* **spettro residuo** σr(T): contiene i λ tali che T - λI è iniettivo ma non suriettivo e la sua immagine non è densa in H.
* **spettro continuo** σc(T): contiene i λ tali che T - λI è iniettivo, non suriettivo ma con immagine densa in H.

**Proprietà dello spettro:**

* σ(T) è un sottoinsieme chiuso e limitato del piano complesso.
* σ(T) non è vuoto se F = C.
* σ(T\*) è il coniugato complesso di σ(T).
* T è invertibile se e solo se 0 ∉ σ(T).

### Autospazi

Dato un operatore lineare limitato T ∈ B(H), l'**autospazio** ET(λ) relativo a λ ∈ F è il nucleo di T - λI, ovvero l'insieme di tutti i vettori x in H tali che Tx = λx.

**Proprietà degli autospazi:**

* Gli autospazi corrispondenti ad autovalori distinti sono in somma diretta.
* Se due operatori commutano, l'immagine e gli autospazi di uno sono invarianti per l'altro.
* Gli autospazi di un operatore normale corrispondenti ad autovalori distinti sono ortogonali.

## Riepilogo degli Operatori Compatti

Questo riassunto fornisce una panoramica degli operatori compatti basandosi sulle informazioni presentate in "10_operatoricompatti_handout.pdf".

### Definizione e Proprietà Fondamentali

**Definizione:**

Siano X e Y spazi normati. Un operatore lineare limitato T da X a Y (T ∈ B(X, Y)) è detto **compatto** se, per ogni successione limitata (xn) in X, la successione delle immagini (Txn) in Y ha una sottosuccessione convergente.

**Proprietà:**

* Un operatore T ∈ B(X, Y) è compatto se e solo se l'immagine T(A) di ogni sottoinsieme limitato A di X è un sottoinsieme compatto di Y.
* L'immagine di un operatore compatto è sempre separabile.
* Ogni operatore di rango finito (cioè con immagine di dimensione finita) è compatto.
* Se X ha dimensione finita oppure Y ha dimensione finita, allora ogni operatore lineare limitato da X a Y è compatto.
* La composizione di operatori lineari limitati è compatta se almeno uno degli operatori è compatto.
* L'operatore identità su uno spazio normato di dimensione infinita non è compatto.
* Un isomorfismo tra spazi normati non può essere compatto se lo spazio di partenza ha dimensione infinita.

### K(X, Y) come Spazio Vettoriale e Spazio di Banach

* L'insieme degli operatori compatti da X a Y, denotato con K(X, Y), forma un sottospazio vettoriale di B(X, Y).
* Se Y è uno spazio di Banach, allora K(X, Y) è un sottoinsieme chiuso di B(X, Y) rispetto alla norma operatoriale.
* Di conseguenza, se Y è uno spazio di Banach, allora K(X, Y) stesso è uno spazio di Banach rispetto alla norma operatoriale.
* Un operatore T ∈ B(X, Y) è compatto se e solo se esiste una successione di operatori di rango finito in B(X, Y) che converge a T nella norma operatoriale.

### Esempi Notevoli

Il documento fornisce alcuni esempi notevoli di operatori compatti e non compatti:

* **Operatore di shift su ℓ2:** Né l'operatore di shift verso sinistra né il suo aggiunto sono compatti.
* **Operatori di moltiplicazione su ℓ2:** Un operatore di moltiplicazione su ℓ2 è compatto se e solo se la sequenza dei moltiplicatori converge a zero.
* **Operatori integrali:** Gli operatori integrali con nucleo integrale in L2 o C sono compatti.

### Operatori Compatti in Spazi di Hilbert

* Per ogni operatore compatto T da uno spazio normato X a uno spazio di Hilbert H, esiste una successione di operatori di rango finito da X a H che converge a T nella norma operatoriale.
* Un operatore T da uno spazio normato X a uno spazio di Hilbert H è compatto se e solo se esiste una successione di operatori di rango finito da X a H che converge puntualmente a T.
* Un operatore tra spazi di Hilbert ha rango finito se e solo se il suo aggiunto ha rango finito, e in tal caso le dimensioni delle loro immagini coincidono.

## Riepilogo della Teoria di Fredholm e dello Spettro di Operatori Compatti

### Teoria di Fredholm

Consideriamo H uno spazio di Hilbert su un campo F, e denotiamo con I l'operatore identità su H.  Un operatore della forma I - K, dove K è un operatore compatto su H, è chiamato **perturbazione compatta dell'identità**.

**Risultati Importanti:**

* Se T è una perturbazione compatta dell'identità su H, allora l'immagine di T è un sottospazio chiuso di H. Di conseguenza, H si decompone nella somma diretta dei sottospazi ortogonali Ker(T*) e Im(T), dove T* è l'operatore aggiunto di T.
* Per un operatore T che è una perturbazione compatta dell'identità, le seguenti affermazioni sono equivalenti: T è iniettivo, T è suriettivo, T è biunivoco, T è un isomorfismo.
*  **Alternativa di Fredholm:**  Data una perturbazione compatta dell'identità T su H, è possibile solo una di queste due opzioni: o l'equazione Tx = y ha un'unica soluzione x in H per ogni y in H, oppure l'equazione omogenea Tx = 0 ha soluzioni non banali (cioè Ker(T) ≠ {0}).

### Spettro di Operatori Compatti

**Proprietà dello Spettro di un Operatore Compatto T su H:**

* Se H ha dimensione infinita, allora 0 appartiene allo spettro di T .
* Per ogni autovalore λ diverso da zero di T, la dimensione dell'autospazio Ker(T - λI) è finita e strettamente positiva.
* Ogni elemento dello spettro di T diverso da zero è un autovalore di T.
* Per ogni numero reale t > 0,  il numero di autovalori di T con modulo maggiore o uguale a t è finito.
* Lo spettro di T è un insieme al più numerabile. Se lo spettro è infinito, l'unico suo punto di accumulazione è 0.

**Operatori Autoaggiunti:**

* Se T è un operatore autoaggiunto, allora almeno uno tra  ∥T∥<sub>op</sub> e -∥T∥<sub>op</sub> appartiene allo spettro di T.

**Operatori Normali:**

Un operatore lineare limitato T su uno spazio di Hilbert complesso H si dice **normale** se  TT* = T*T, dove T* è l'operatore aggiunto di T.

* Se T è un operatore compatto e normale su uno spazio di Hilbert complesso, allora:
    * Lo spettro di T è un insieme finito o numerabile con l'unico possibile punto di accumulazione in 0.
    * Per ogni autovalore  λ diverso da zero, la dimensione dell'autospazio corrispondente (denotata con  ET(λ)) è finita.

### Teorema Spettrale per Operatori Compatti e Normali

**Teorema (Versione per Operatori Compatti e Normali):** Sia T un operatore compatto e normale su uno spazio di Hilbert H. Se H è uno spazio di Hilbert reale, assumiamo che T sia autoaggiunto (T=T*). Allora:

1.  Vale la decomposizione: Ker(T) = (⋃<sub>λ∈σ(T)\{0}</sub> ET(λ))<sup>⊥</sup>,  dove  ET(λ) = Ker(T - λI) .

2.  Esiste una base ortonormale B di H con le seguenti proprietà:
    *  B è l'unione di basi ortonormali degli autospazi  ET(λ)  per λ diverso da zero.
    *  Ogni elemento x di H si può scrivere in modo unico come combinazione lineare (infinita) degli elementi di B, con i coefficienti dati dai prodotti scalari di x con gli elementi di B.
    *  L'azione di T su x si ottiene moltiplicando ogni elemento di B per il corrispondente autovalore e sommando i risultati. La convergenza di questa serie è incondizionata.

**Corollario:** Se, nelle ipotesi del teorema precedente, H è anche separabile, allora H ammette una base ortonormale composta da autovettori di T.
